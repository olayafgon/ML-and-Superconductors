```python
import os
import sys
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import time

sys.path.append('./..')
```

```python
# Directorio donde se encuentran los archivos
directorio = r"D:\tfg\data\data_raw\BCC"
```

```python
%%time

# Función para procesar cada archivo y extraer la información
def procesar_archivo(archivo):
    
    with open(archivo, 'r') as f:
        lines = f.readlines()

    # Verificar si el archivo está vacío o si no se encuentra la cabecera
    if not lines or len(lines) < 6 or not lines[4].startswith(" "):
        print(f'· Error: {archivo}')
        return None

    # Obtener fórmula y código ICSD
    cabecera = lines[4].split('_')
    formula = cabecera[0]
    codigo_icsd = int(cabecera[2].strip())


    # Obtener energía de Fermi
    energia_fermi = float(lines[5].split()[3])

    # Obtener densidades de estados dentro del rango de energía especificado
    densidades_de_estados = []
    densidades_de_estados = []
    for line in lines[5:5001+5]:
        datos = line.split()
        densidad_de_estado = float(datos[1])
        densidades_de_estados.append(densidad_de_estado)

    return [formula, codigo_icsd, energia_fermi] + densidades_de_estados

# Lista para almacenar los datos de cada archivo
datos_totales = []

# Iterar sobre cada archivo en el directorio
for archivo in os.listdir(directorio):
    if archivo.endswith('.txt'):  # Solo procesar archivos de texto
        datos_archivo = procesar_archivo(os.path.join(directorio, archivo))
        if datos_archivo is not None:
            datos_totales.append(datos_archivo)
```

```python
# Definir nombres de columnas para el DataFrame
columnas = ['Formula', 'Codigo ICSD', 'Energia Fermi'] + [f'E_{np.round(-30 + i * (75/5000),4)} eV' for i in range(5001)]

# Crear DataFrame
df = pd.DataFrame(datos_totales, columns=columnas)

df
```
